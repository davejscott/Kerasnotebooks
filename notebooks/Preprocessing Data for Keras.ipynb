```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

# Example data

### Generate data below
#### simulation of two different categories (young / old) being tested on a hypothetical new drug

```python
for i in range(1000):
    random_younger = randint(13, 64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
    
for i in range(50):
    random_youinger = randint(13, 64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
```

```python
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

# 0's indicate no side effects. 1 indicate side effects

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
```

# Note, Keras needs labels and samples in a np.array

# Neural net may struggle due to the poor variation in data. Let's scale using MinMaxScaler

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform((train_samples).reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

# ages in the scaled 0 to 1. Net will make better predictions
